```python
# Get Season Games until now
import datetime
def get_current_season():
    current_season = datetime.datetime.strptime("12-22-2020", '%m-%d-%Y')
    today = datetime.date.today()
    
    boxscore = Boxscores(current_season, today)

    d = boxscore.__dict__

    d_cleaned = d['_boxscores']
    d_cleaned = {k: v for k, v in d.items() if len(v) is not 0}

    d_vals = list(d_cleaned.values())

    from collections import defaultdict

    d = defaultdict(list)

    for outer_lst in d_vals:
        for i in outer_lst:
            for k, v in i.items():
                d[k].append(v)
    
    df_boxscore = pd.DataFrame.from_dict(d)

    return df_boxscore
```

```python
df_current_season = get_current_season()
df_current_season
```
